```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
file =r"C:\Users\bbggw\Downloads\m5_survey_data_technologies_normalised.csv"
```

```python
df= pd.read_csv(file)
```

```python
df
```

```python
LangVC= df['LanguageWorkedWith'].value_counts(sort=True, ascending=False, dropna=True)
```

```python
Lang =LangVC[:10].index
Lang
```

```python
fig, ax = plt.subplots()


ax.barh(Lang, LangVC[:10], align='center')
ax.set_ylabel('Languages')

ax.set_xlabel('Number of respondents who use the Language')
ax.set_title('Top Ten Languages Worked With in Current Tech')

plt.show()
```

```python
FutL= df['LanguageDesireNextYear'].value_counts(sort=True, ascending=False, dropna=True)
topF =FutL[:10].index
fig, ax = plt.subplots()


ax.barh(topF, FutL[:10], align='center')
ax.set_ylabel('Languages')

ax.set_xlabel('Number of respondents who wish to use this Language')
ax.set_title('Top Ten Languages Desired to Work With in Tech')

plt.show()
```

```python
#database trends
dbc= df['DatabaseWorkedWith'].value_counts(sort=True, ascending=False, dropna=True)
dbcl =dbc[:10].index
fig, ax = plt.subplots()

plt.bar(dbcl, dbc[:10])

# Rotate x-axis labels
plt.xticks(rotation=65)

# Add labels and title
plt.xlabel('Databases')
plt.ylabel('Respondent Count')
plt.title('Current Databases')


plt.show()
```

```python
df4= df['DatabaseDesireNextYear'].value_counts(sort=True, ascending=False, dropna=True)
df4l =df4[:10].index
fig, ax = plt.subplots()

plt.bar(df4l, df4[:10])

# Rotate x-axis labels
plt.xticks(rotation=65)

# Add labels and title
plt.xlabel('Databases')
plt.ylabel('Respondent Count')
plt.title('Future Databases')


plt.show()
```
